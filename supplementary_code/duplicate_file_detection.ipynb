```python
import os
from nltk.tokenize import word_tokenize
from scipy.spatial.distance import jaccard
```

```python
#file_path="tensors/x86_benign"
file_path="../original_dataset/gdl_pb_formatted/s390x"
similar_threshold=0.99
duplicate_path=f"../output/duplicate/s390x_{similar_threshold}.txt"
```

```python
def detect_similar_files(directory, threshold):
    files = os.listdir(directory)
    #print(files)
    file_tokens = {}
    num=0

    for file in files:
        with open(os.path.join(directory, file), 'r') as f:
            code = f.read()
            tokens = set(code.split())
            file_tokens[file] = tokens
        #print(f"File: {file}, tokens: {file_tokens[file]}")    

    removed_files = set()
    for i in range(len(files)):
        for j in range(i + 1, len(files)):
            #print(1)
            file1, file2 = files[i], files[j]
            if file1 not in removed_files and file2 not in removed_files:
                tokens1=file_tokens[file1]
                tokens2=file_tokens[file2]
                intersection=len(tokens1&tokens2)
                union=len(tokens1|tokens2)
                similarity=intersection/union
                if similarity > threshold:
                    removed_files.add(file2)
                    print(f"{file1}<--->{file2}")

    return removed_files                
```

```python
def delete_duplicate_files(directory, duplicate_files):
    for file in duplicate_files:
        file_path = os.path.join(directory, file)
        if os.path.exists(file_path):
            os.remove(file_path)
            print(f"{file_path} has been removed")
        else:
            print(f"{file_path} does not exist")
```

```python
def main():
    removed_files=detect_similar_files(file_path, similar_threshold)
    sorted_removed_files=sorted(removed_files,key=lambda x:x.split('-')[-1])
    formatted_removed_files=[file.replace("_formatted","") for file in sorted_removed_files]
    print(f"The total number of files to remove is: {len(removed_files)}")
    #print("The following files will be removede")
    # for item in removed_files:
    #     print(item)
    with open(duplicate_path, "w") as file:
        for item in formatted_removed_files:
            file.write(f"{item}\n")    
    #delete_duplicate_files(file_path, removed_files)        

if __name__ == "__main__":
    main()    
```
