```python
import os
```

```python
#x86_folder = "data/parallel_dataset_evaluation/ppc32-x86/x86-zlib"
#src_folder = "data/parallel_dataset_evaluation/ppc32-x86/ppc32-zlib"

x86_folder="../original_dataset/translated_x86_pb/x86_reference"
src_folder="../original_dataset/translated_x86_pb/i386"

# Get the list of files in the x86 folder
x86_files = os.listdir(x86_folder)
# Get the list of files in the src folder
src_files = os.listdir(src_folder)
```

```python
files_to_delete = []

for file in src_files:
    # Check if the file exists in the x86 folder
    file_path = os.path.join(src_folder, file)
    with open(file_path, 'r', encoding='utf-8') as filename:
        file_content = filename.read()
    if file_content == "":
        files_to_delete.append(file_path)
    if file not in x86_files:
        # Mark the file for deletion from the src folder if it doesn't exist in x86
        files_to_delete.append(file_path)

# Delete the files marked for deletion from the src folder
for file_path in files_to_delete:
    if os.path.exists(file_path):
        os.remove(file_path)

# Iterate over the files in the x86 folder
for file in x86_files:
    file_path = os.path.join(x86_folder, file)
    with open(file_path, 'r', encoding='utf-8') as filename:
        file_content = filename.read()
    if file_content == "":
        os.remove(file_path)
    # Check if the file exists in the src folder
    if file not in src_files:
        # Delete the file from the x86 folder if it doesn't exist in src
        file_path = os.path.join(x86_folder, file)
        if os.path.exists(file_path):
            os.remove(file_path)

for file in os.listdir(src_folder):
    print(file)
```
