```python
def read_vocab(file_path):
    vocab_dict={}
    with open(file_path,'r') as text:
        lines=text.readlines()
        for line in lines:
            token=line.split()[0]
            occurence=line.split()[1]
            vocab_dict[token]=occurence
    return vocab_dict
```

```python
def read_file(file_path):
    vocab_dict={}
    with open(file_path,'r') as text:
        token_count=0 #Only calculate the first 500,000 tokens. 
        file_threshold=550000
        lines=text.readlines()
        for line in lines:
            tokens=line.split()
            for token in tokens:
                token_count+=1
                if token_count>file_threshold:
                    return vocab_dict
                if token in vocab_dict:
                    vocab_dict[token]+=1
                else:
                    vocab_dict[token]=1
    return vocab_dict            
```

```python
def main():
    vocab_path=f"../original_dataset/ir_formatted/BPE/i386ir_malware_vocab.txt"
    file_path=f"../Malware_detection_files/tensors/BPE/i386ir/i386ir_malware/malware_i386_filtered-VirusShare_92cf74d4c4065c92ff32bc027a9ae039.txt"
    
    vocab_dict=read_vocab(vocab_path)
    vocab_set=set()

    for key in vocab_dict:
        vocab_set.add(key)

    file_dict=read_file(file_path)
    file_set=set()
    for key in file_dict:
        file_set.add(key)
    unique_set=file_set.difference(vocab_set)
    print(unique_set)

if __name__ == '__main__':
    main()     
```
