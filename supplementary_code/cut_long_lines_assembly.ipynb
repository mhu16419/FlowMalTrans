```python
import os
```

```python
ISA="m68k"
#ISA="i386"
input_folder=f"../original_dataset/gdl_pb_formatted/{ISA}"
output_folder=f"../original_dataset/gdl_pb_formatted/{ISA}_cutted"
```

```python
def main():

    os.makedirs(output_folder,exist_ok=True)
    for filename in os.listdir(input_folder):
        with open(os.path.join(input_folder,filename)) as f:
            content=f.read()
        lines=content.split("\n")
        output_lines=[]
        for line in lines:
            tokens = line.split()
            
            # Check if the line needs to be split
            if len(tokens) >= 50:
                new_line = []
                for token in tokens:
                    new_line.append(token)
                    if len(new_line) >= 25 and token == '<E>':
                        output_lines.append(" ".join(new_line))
                        new_line = []
                if new_line:
                    output_lines.append(" ".join(new_line))
            else:
                output_lines.append(line)
        
        # Write the output lines to a new file in the output folder
        output_filename = f"{filename}"
        with open(os.path.join(output_folder, output_filename), 'w') as file:
            file.write("\n".join(output_lines))    

if __name__ == '__main__':
    main()    
```
