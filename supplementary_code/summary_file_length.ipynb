### This file output the txt file to make sure each line is less than 200. 

```python
import os
import glob
```

```python
def load_txt(file_path):
    with open(file_path,'r',encoding='utf-8') as file:
        text=file.read()
    return text    
```

```python
def main():
# Specify the folder path
    ISA="x86"
    #folder_path = f"../Malware_detection_files/tensors/BPE/{ISA}/{ISA}_benign"
    folder_path = f"../Malware_detection_files/tensors/BPE_300_20240815/{ISA}/{ISA}_malware"
    folder_path = f"../original_dataset/gdl_pb_formatted/{ISA}_malware"
    file_names=os.listdir(folder_path)
    file_paths=[os.path.join(folder_path,name) for name in file_names]
    file_lengths=[]

    for file in file_paths:
        file_len=0
        text=load_txt(file)
        lines=text.split('\n')
        for line in lines:
            #print(f"line length is {len(line.strip().split())}")
            file_len+=len(line.strip().split())

        file_name=file.split('/')[-1]
        file_lengths.append((file_name,file_len))
        #print(f"File {file_name} length is:{file_len}")
    sorted_file_lengths=sorted(file_lengths,key=lambda x:x[1],reverse=False)

    # Call the function to analyze the files
    
    for file_name,length in sorted_file_lengths:
        print(f"{file_name}: {length}")


if __name__ == '__main__':
    main()   
```
