This script calculates the 95% quantile frequency of the vocabulary list. 

```python
import math
```

```python
#vocab_path="bpe_process/codes_vocab/x86_benign_vocab.txt"
vocab_path="data/processed/arm64-x86/vocab.arm64-x86"
```

```python
def find_95_quantile_frequency(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()

    tokens = []
    for line in lines:
        token, frequency = line.strip().split(' ')
        tokens.append((token, int(frequency)))

    tokens.sort(key=lambda x: x[1], reverse=True)

    total_tokens = len(tokens)
    quantile_index = math.ceil(0.95 * total_tokens) - 1

    return total_tokens, tokens[quantile_index][1]

# Example usage

total_tokens,frequency = find_95_quantile_frequency(vocab_path)
print(f"The total number of tokens in the vocabulary is: {total_tokens}")
print(f"The frequency of the token at the 95% quantile is: {frequency}")     
```
