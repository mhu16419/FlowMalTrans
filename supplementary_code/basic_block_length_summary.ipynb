```python
import os
```

```python
vocab_path="data/processed/arm64-x86_old/vocab.arm64-x86"
```

```python
def calculate_divergence(vocab_list):
    total_freq=sum(vocab_list.values())
    vocab_size=len(vocab_list)
    uniform_prob=1/vocab_size
    divergence=0
    for freq in vocab_list.values():
        prob=freq/total_freq
        divergence+=abs(prob-uniform_prob)

    return divergence                  
```

```python
def main():
    with open(vocab_path,'r') as file:
        lines=file.readlines()
    vocab_list={}
    selected_size=len(lines)//20
    for line in lines[:selected_size]:
        token, freq=line.strip().split()
        vocab_list[token]=int(freq)    
  
    divergence=calculate_divergence(vocab_list)
    print(f"divergence: {divergence:.4f}")

if __name__ == "__main__":
    main()    
```
