This file extracts the opcode list from the .gdl files. 

```python
import os
import glob # folder operations
```

```python
def main():
    ISA='arm32'
    # Use glob to find all '.gdl' files recursively in the current folder and sub-folders
    input_path = glob.glob(f"../original_dataset/gdl_pb_unformatted/{ISA}/*")
    output_path=f"../original_dataset/opcode_operand/{ISA}"
    os.makedirs(output_path, exist_ok=True)

    opcode_dict = {}
    # Process each file
    for gdl_file_path in input_path:
        with open(gdl_file_path, 'r') as f:
            lines = f.readlines()
            for line in lines:
                token = line.strip().split()[0]
                opcode_dict[token] = opcode_dict.get(token, 0) + 1

    # Sort the dictionary by occurrence count in descending order
    sorted_opcode_dict = sorted(opcode_dict.items(), key=lambda x: x[1], reverse=True)

    # Print the dictionary line by line
    for token, count in sorted_opcode_dict:
        print(f"{token}: {count}")

    # Write the output to a file
    output_file = os.path.join(output_path, "opcode_counts.txt")
    with open(output_file, 'w') as f:
        for token, count in sorted_opcode_dict:
            f.write(f"{token} {count}\n")

if __name__ == '__main__':
    main()
```
